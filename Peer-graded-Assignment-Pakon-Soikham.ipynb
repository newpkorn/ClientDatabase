## Task 2: Implement a stored procedure called PeakHours
Step one: Write a SQL CREATE PROCEDURE query for PeakHours

Use HOUR to extract the hour part from the BookingSlot.

Use COUNT on hour to count the number of bookings.

Use GROUP BY on booking hour.

Use ORDER BY on the number of bookings in descending order.

Step two: Run the stored procedure query by invoking execute module on the cursor.

Step three: Invoke callproc to call the stored procedure.

Step four: Fetch the results in a variable called dataset.

Step five: Extract the names of the columns.

Step six: Print the names of the columns.

Step seven: Print the sorted data using for loop.

```python
# Step 1: Write a SQL CREATE PROCEDURE query for PeakHours
"""
CREATE PROCEDURE PeakHours()
BEGIN
    SELECT HOUR(BookingSlot) AS BookingHour, COUNT(*) AS NumberOfBookings
    FROM Bookings
    GROUP BY BookingHour
    ORDER BY NumberOfBookings DESC;
END;
"""


# Step 2: Run the stored procedure query by invoking execute on the cursor
try:
    cursor.execute("DROP PROCEDURE IF EXISTS PeakHours")
    cursor.execute("""
        CREATE PROCEDURE PeakHours()
        BEGIN
            SELECT HOUR(BookingSlot) AS BookingHour, COUNT(*) AS NumberOfBookings
            FROM Bookings
            GROUP BY BookingHour
            ORDER BY NumberOfBookings DESC;
        END;
    """)
    print("Stored procedure PeakHours created successfully.")
except Error as err:
    print(f"Error: {err}")

# Step 3: Invoke callproc to call the stored procedure
try:
    cursor.callproc('PeakHours')
    print("Stored procedure PeakHours called successfully.")
except Error as err:
    print(f"Error: {err}")

# Step 4: Fetch the results in a variable called dataset
dataset = []
for result in cursor.stored_results():
    dataset = result.fetchall()

# Step 5: Extract the names of the columns
columns = [desc[0] for desc in result.description]

# Step 6: Print the names of the columns
print(" | ".join(columns))

# Step 7: Print the sorted data using a for loop
for row in dataset:
    print(" | ".join(map(str, row)))


```

## Task 3: Implement a stored procedure GuestStatus
Step one: Write a SQL CREATE PROCEDURE query for GuestStatus.

Step two: Combine the guest’s first and last name from the booking column using CONCAT. 

Step three: Use CASE to implement the following statuses for each guest’s order:

If the Role in the Employee table is Manager or Assistant Manager then the guest’s order status is Ready to pay

If the Role in the Employee table is Head Chef then the status is Ready to serve

If the Role in the Employee table is Assistant Chef then the status is Preparing Order

If the Role in the Employee table is Head Waiter then the status is Order served

Step four: LEFT JOIN Bookings table with Employees ON EmployeeID

Step five: Run the stored procedure query by invoking execute module on the cursor.

Step six: Invoke callproc to call the stored procedure.

Step seven: Fetch the results in a variable called dataset.

Step eight: Extract the names of the columns.

Step nine: Print the names of the columns.

Step ten: Print the sorted data using for loop.

Step eleven: Close the connection to return it back to the pool.

```python
# Step 1: Write a SQL CREATE PROCEDURE query for GuestStatus
"""
CREATE PROCEDURE GuestStatus()
BEGIN
    SELECT 
        CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) AS GuestName,
        CASE 
            WHEN Employees.Role = 'Manager' OR Employees.Role = 'Assistant Manager' THEN 'Ready to pay'
            WHEN Employees.Role = 'Head Chef' THEN 'Ready to serve'
            WHEN Employees.Role = 'Assistant Chef' THEN 'Preparing Order'
            WHEN Employees.Role = 'Head Waiter' THEN 'Order served'
            ELSE 'Unknown status'
        END AS OrderStatus
    FROM Bookings
    LEFT JOIN Employees ON Bookings.EmployeeID = Employees.EmployeeID;
END;
"""

"""
Step 2: Combine the guest’s first and last name using CONCAT
This step is already covered in the SELECT statement within the CREATE PROCEDURE query above.

Step 3: Use CASE to implement the statuses
The CASE statement in the SELECT clause checks the role of the employee and assigns the corresponding status.

Step 4: LEFT JOIN the Bookings table with Employees ON EmployeeID
This is also included in the SELECT statement within the CREATE PROCEDURE query.

Step 5: Run the stored procedure query by invoking execute on the cursor
"""

from mysql.connector import pooling, Error, PoolError  # Ensure PoolError is imported

try:
    cursor.execute("DROP PROCEDURE IF EXISTS GuestStatus")
    cursor.execute("""
        CREATE PROCEDURE GuestStatus()
        BEGIN
            SELECT 
                CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) AS GuestName,
                CASE 
                    WHEN Employees.Role = 'Manager' OR Employees.Role = 'Assistant Manager' THEN 'Ready to pay'
                    WHEN Employees.Role = 'Head Chef' THEN 'Ready to serve'
                    WHEN Employees.Role = 'Assistant Chef' THEN 'Preparing Order'
                    WHEN Employees.Role = 'Head Waiter' THEN 'Order served'
                    ELSE 'Unknown status'
                END AS OrderStatus
            FROM Bookings
            LEFT JOIN Employees ON Bookings.EmployeeID = Employees.EmployeeID;
        END;
    """)
    print("Stored procedure GuestStatus created successfully.")

    cursor.callproc('GuestStatus')
    print("Stored procedure GuestStatus called successfully.")

    dataset = []
    for result in cursor.stored_results():
        dataset = result.fetchall()

    columns = [desc[0] for desc in result.description]
    print(" | ".join(columns))

    for row in dataset:
        print(" | ".join(map(str, row)))

except Error as err:
    print(f"Error: {err}")

finally:
    if connection:
        try:
            connection.close()
            print("Connection returned to the pool.")
        except PoolError as pe:
            print(f"Failed to return connection to the pool: {pe}")
```

## Task 2:
Three guests are trying to book dinner slots simultaneously. Get the connections from pool_b and insert the following data in the Bookings table:


TIP: You need to add a connection to connect the third guest.


Guest 1:

First Name: Anees

EmployeeID: 6


Guest 3:

Table Number: 12

First Name: Jay

Return all the connections back to the pool. 


TIP: The pool size is two. However, you have three connected users. You can only return two connections. Returning a third connection will raise a PoolError. Use try-except to print the error message.  


```python
from mysql.connector import pooling, Error, PoolError

# Define database configurations
dbconfig = {
    "database": "little_lemon_db",
    "user": "root",
    "password": ""
}

# Create the connection pool
try:
    pool_b = pooling.MySQLConnectionPool(
        pool_name="pool_b",
        pool_size=3,  # Adjusted pool size to handle 3 concurrent connections
        **dbconfig
    )
    print("Connection pool 'pool_b' created successfully.")
except Error as err:
    print(f"Error: {err}")

# Define booking data
bookings = [
    (8, 'Anees', 'Java', '18:00', 6),
    (5, 'Bald', 'Vin', '19:00', 6),
    (12, 'Jay', 'Kon', '19:30', 6)
]

connections = []
try:
    # Obtain connections from the pool
    for i in range(len(bookings)):
        connection = pool_b.get_connection()
        connections.append(connection)

    # Insert data into the Bookings table
    for index, connection in enumerate(connections):
        try:
            cursor = connection.cursor()
            query = """
                INSERT INTO Bookings (TableNo, GuestFirstName, GuestLastName, BookingSlot, EmployeeID)
                VALUES (%s, %s, %s, %s, %s)
            """
            cursor.execute(query, bookings[index])
            connection.commit()
            print(f"Booking data for guest {index + 1} inserted successfully.")
        except Error as err:
            print(f"Error executing query for guest {index + 1}: {err}")
        finally:
            cursor.close()
            connection.close()  # Properly close and return to the pool

except Error as err:
    print(f"Error: {err}")
finally:
    # Ensure all connections are properly closed
    for connection in connections:
        if connection:
            try:
                connection.close()  # Properly close connection
                print("Connection closed and returned to the pool.")
            except PoolError as pe:
                print(f"Failed to return connection to the pool: {pe}")
```

## Task 4:
Create a stored procedure named BasicSalesReport that returns the following statistics: 

Total sales

Average sale

Minimum bill paid

Maximum bill paid

```python
from mysql.connector import pooling, Error

# Define database configurations
dbconfig = {
    "database": "little_lemon_db",
    "user": "root",
    "password": ""
}

# Create the connection pool
try:
    pool_b = pooling.MySQLConnectionPool(
        pool_name="pool_b",
        pool_size=2,
        **dbconfig
    )
    print("Connection pool 'pool_b' created successfully.")
except Error as err:
    print(f"Error: {err}")

# Create and call the stored procedure
try:
    connection = pool_b.get_connection()
    cursor = connection.cursor()

    # Drop the existing procedure if it exists
    cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport")

    # Create the stored procedure
    cursor.execute("""
    CREATE PROCEDURE BasicSalesReport()
    BEGIN
        SELECT 
            SUM(BillAmount) AS TotalSales,
            AVG(BillAmount) AS AverageSale,
            MIN(BillAmount) AS MinimumBillPaid,
            MAX(BillAmount) AS MaximumBillPaid
        FROM Orders;
    END;
    """)
    print("Stored procedure BasicSalesReport created successfully.")

    # Call the stored procedure
    cursor.callproc('BasicSalesReport')
    print("Stored procedure BasicSalesReport called successfully.")

    # Fetch and print the results
    for result in cursor.stored_results():
        report = result.fetchall()
        for row in report:
            print(f"Total Sales: {row[0]}")
            print(f"Average Sale: {row[1]}")
            print(f"Minimum Bill Paid: {row[2]}")
            print(f"Maximum Bill Paid: {row[3]}")

    cursor.close()
    connection.close()
    print("Connection returned to the pool.")

except Error as err:
    print(f"Error: {err}")
```

## Task 5:
Little Lemon needs to display the next three upcoming bookings from the Bookings table on the kitchen screen to notify their chefs which orders are due next. To complete this task, carry out the following steps:

Get a connection from the pool.

Create a buffered cursor.

Combine the data from the Bookings and the Employee tables. Sort the retrieved records in ascending order. Then display the information of the first three guests. 

Returned the connection back to the pool.


The output should be as follows:

[BookingSlot]

[Guest_name]

[Assigned to: Employee Name [Employee Role]]



```python
from mysql.connector import pooling, Error

# Define database configurations
dbconfig = {
    "database": "little_lemon_db",
    "user": "root",
    "password": ""
}

# Create the connection pool
try:
    pool_b = pooling.MySQLConnectionPool(
        pool_name="pool_b",
        pool_size=2,
        **dbconfig
    )
    print("Connection pool 'pool_b' created successfully.")
except Error as err:
    print(f"Error: {err}")

# Retrieve the next three upcoming bookings
try:
    connection = pool_b.get_connection()
    cursor = connection.cursor(buffered=True)

    # Query to get the next three upcoming bookings
    query = """
    SELECT 
        Bookings.BookingSlot, 
        CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) AS GuestName,
        CONCAT(Employees.Name, ' [', Employees.Role, ']') AS EmployeeInfo
    FROM Bookings
    LEFT JOIN Employees ON Bookings.EmployeeID = Employees.EmployeeID
    ORDER BY Bookings.BookingSlot ASC
    LIMIT 3;
    """

    cursor.execute(query)

    # Fetch and print the results
    print("Upcoming Bookings:")
    for (booking_slot, guest_name, employee_info) in cursor:
        print(f"[{booking_slot}]")
        print(f"[{guest_name}]")
        print(f"[Assigned to: {employee_info}]")
        print()

    cursor.close()
    connection.close()
    print("Connection returned to the pool.")

except Error as err:
    print(f"Error: {err}")
```
